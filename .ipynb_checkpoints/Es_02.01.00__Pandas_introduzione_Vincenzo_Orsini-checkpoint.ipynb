```python
import matplotlib.pyplot as plt
import pandas as pd
import wget
```

```python
wget.download('https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-regioni/dpc-covid19-ita-regioni.csv', out='data')
```

```python
df = pd.read_csv('dpc-covid19-ita-regioni.csv')
```

```python
df
```

```python
df.columns
```

```python
df_campania = df.loc[df["denominazione_regione"]=="Campania"]
df_campania
```

Totale Positivi in Campania in funzione della data:

```python
plt.figure(figsize=(20,15))
plt.plot(df_campania["data"], df_campania["totale_positivi"])
plt.ylabel("Totale Positivi")
plt.xlabel("Data")
plt.show()
```

Positivi Giornalieri in funzione della data

```python
plt.figure(figsize=(20,15))
plt.plot(df_campania["data"], df_campania["totale_positivi"]-df_campania["totale_positivi"].shift(1), label="Con metodo shift")
plt.plot(df_campania["data"], df_campania["nuovi_positivi"], label="Con dati tabulati")
plt.ylabel("Positivi Giornalieri")
plt.xlabel("Data")
plt.legend()
plt.show()
```

Media mobile a 7 giorni dei positivi giornalieri in Campania in funzione della data

```python
plt.figure(figsize=(20,15))
plt.plot(df_campania["data"], df_campania["totale_positivi"].rolling(window=7).mean())
plt.ylabel("Totale Positivi")
plt.xlabel("Data")
plt.show()
```

```python
df_recent=df.loc[df["data"]=="2022-03-21T17:00:00"].copy().reset_index(drop=True)
df_recent[["denominazione_regione", "deceduti"]]
df_recent=df_recent.sort_values(by=['deceduti'], ascending=False)

fig, ax = plt.subplots(figsize=(20,15))
ax.bar(df_recent["denominazione_regione"], df_recent["deceduti"])
ax.set_xlabel("regione")
ax.set_ylabel("deceduti")
plt.title("21-03-2022 ore 00:00")
plt.xticks(rotation="vertical")
plt.show()
```

Distribuzione del numero di positivi giornalieri rispetto al numero di tamponi fatti in funzione della data

```python
plt.figure(figsize=(20,15))
plt.plot(df_campania["data"], (df_campania["totale_positivi"]-df_campania["totale_positivi"].shift(1))/(df_campania["tamponi"]-df_campania["tamponi"].shift(1)))
plt.ylabel("Rateo Positivi giornalieri/nuovi tamponi")
plt.xlabel("Data")
plt.xticks(rotation="vertical")
plt.show()
```

```python
wget.download('https://www.football-data.co.uk/mmz4281/2021/I1.csv', out='SerieA')
```

```python
slave = SerieA.HomeTeam.copy().iloc[0:30]
slave.duplicated()
slave = slave.drop_duplicates(keep="first")
slave
```

```python
SerieA = pd.read_csv("SerieA")
SerieA[["HomeTeam", "AwayTeam", "FTR"]]
Table = []
for i in slave:
    ds=SerieA[SerieA["HomeTeam"]==i]
    da=SerieA[SerieA["AwayTeam"]==i]
    Points = 0
    for j in ds.FTR:
        if j == "H":
            Points +=3
        elif j == "D":
            Points +=1
        elif j == "A":
            Points +=0
    for k in da.FTR:
        if k == "H":
            Points +=0
        elif k == "D":
            Points +=1
        elif k == "A":
            Points +=3
    Table.append([i, Points])
Leaderboard = pd.DataFrame(Table, columns=["Club", "Points"])
Leaderboard.sort_values("Points", ascending=False, ignore_index=True)
```
